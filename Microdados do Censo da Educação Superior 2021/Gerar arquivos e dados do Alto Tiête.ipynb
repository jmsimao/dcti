```python
import numpy as np
import pandas as pd
```

```python
arquivo1=False

if (arquivo1):
    ## Arquivo 1
    arquivos={'entrada':'MICRODADOS_CADASTRO_IES_2021.zip', 'saida':'MICRODADOS_ED_SUP_IES_2021_ALTO_TIETE.csv'}
    atr_selecao='NO_MUNICIPIO_IES'
    drop_columns=['NU_ANO_CENSO','NO_REGIAO_IES','CO_REGIAO_IES','NO_UF_IES','SG_UF_IES','CO_UF_IES','IN_CAPITAL_IES',
                  'NO_MESORREGIAO_IES','CO_MESORREGIAO_IES',
                  'DS_ENDERECO_IES','DS_NUMERO_ENDERECO_IES','DS_COMPLEMENTO_ENDERECO_IES','NO_BAIRRO_IES','NU_CEP_IES'
                 ]
else:
    ## Arquivo 2
    arquivos={'entrada':'MICRODADOS_CADASTRO_CURSOS_2021.zip', 'saida':'MICRODADOS_CADASTRO_CURSOS_2021_ALTO_TIETE.csv'}
    atr_selecao='NO_MUNICIPIO'
    drop_columns=['NU_ANO_CENSO','NO_REGIAO','CO_REGIAO','NO_UF','SG_UF','CO_UF']
```

```python
ces=pd.read_csv('./Dados/'+arquivos['entrada'],
            sep=';',
            header=0,
            encoding='latin-1',
            low_memory=False,
            #on_bad_lines='skip'
            compression='zip'
            )
```

```python
ces.drop(labels=drop_columns, axis=1, inplace=True)

cidades=['Arujá','Biritiba Mirim','Ferraz de Vasconcelos','Guararema','Guarulhos','Itaquaquecetuba','Mogi das Cruzes',
         'Poá','Salesópolis','Santa Isabel','Suzano']
cesat=ces[(ces[atr_selecao].isin(cidades))]
cesat.reset_index(inplace=True, drop=True, level=0)

cesat
```

```python
#### Eliminação dos atributos e linhas.
def eliminaAtributosELinhasArquivo(max):
    desc.drop(axis=1, labels='Unnamed: 0', inplace=True)
    for i in range(3,max):
        desc.drop(axis=1, labels='Unnamed: '+str(i), inplace=True)

    desc.dropna(how='any', inplace=True)
    desc.drop(axis=0, index=4, inplace=True)

def eliminaAtributosELinhasArquivo2():
    desc.drop(axis=1, labels='Unnamed: 0', inplace=True)
    for i in range(3,24):
        desc.drop(axis=1, labels='Unnamed: '+str(i), inplace=True)

    desc.dropna(how='any', inplace=True)
    desc.drop(axis=0, index=4, inplace=True)

if (arquivo1):
    eliminaAtributosELinhasArquivo(26)

    desc.columns=['ATRIBUTO','DESCRICAO']
    desc.reset_index(inplace=True, level=0, drop=True)
    desc.set_index('ATRIBUTO', drop=True, inplace=True)
    print(desc.loc['CO_REGIAO_IES']['DESCRICAO'])
else:
    eliminaAtributosELinhasArquivo(23)

    desc.columns=['ATRIBUTO','DESCRICAO']
    desc.reset_index(inplace=True, level=0, drop=True)
    desc.set_index('ATRIBUTO', drop=True, inplace=True)
    print(desc.loc['QT_MAT_MOB_ACADEMICA']['DESCRICAO'])
    
```
