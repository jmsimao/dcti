#### Mapa de Calor.
##### https://seaborn.pydata.org/generated/seaborn.heatmap.html
##### https://matplotlib.org/stable/users/explain/colors/colormaps.html

```python
def mapaDeCalor(hpDataFrame, colunas, descricao, colunasAgrupadas) -> pd.DataFrame:
    ## DataFrame, Colunas Originais dos Dados, Descrição dos dados (tabular), Nome das Colunas de Dados Agrupados.
    df = hpDataFrame[colunas]
    tabela=[]

    #for r in range(1, len(nomeColunas)):
    #    print(nomeColunas[r])
    
    for i in df.values:
        for r in range(1, len(descricao)):
            tabela.append([i[0], descricao[r], i[r]])

    novoDF=pd.DataFrame(tabela)
    novoDF.columns=colunasAgrupadas

    #print(novoDF)
    return novoDF

def plotarHeatMap(dados, **kwargs):
    colorMap=kwargs['cmap']

    lw=0
    for kw in kwargs:
        if (kw=='linewidths'):
            lw=kwargs['linewidths']  
        if (kw=='lw'):
            lw=kwargs['lw']  

    plt.figure(figsize=(12,8))
    plt.title(kwargs['titulo'], color='blue', fontdict={'size':16,'weight':'bold'})
    sns.heatmap(dados, lw=lw, annot=True, fmt=".0f", cmap=colorMap)
    plt.xlabel(kwargs['xLabel'],fontdict={'size':14,'weight':'bold'})
    plt.ylabel(kwargs['yLabel'],fontdict={'size':14,'weight':'bold'})
    plt.savefig('./imagens/'+kwargs['titulo'].replace('/','-')+'.jpg', dpi=300, format='jpg')
    plt.show()
```

#### Titulação.

```python
titulo=mapaDeCalor(cesat,
                ['NO_MUNICIPIO_IES','QT_DOC_EX_SEM_GRAD','QT_DOC_EX_GRAD','QT_DOC_EX_ESP','QT_DOC_EX_MEST','QT_DOC_EX_DOUT'],
                ['MUNICÍPIO','SEM GRADUAÇÃO','GRADUAÇÃO','ESPECIALIZAÇÃO','MESTRADO','DOUTORADO'],
                ['MUNICÍPIO','TITULAÇÃO','QTDE']
               )

# Agrupamento de Dados...
titulo=pd.DataFrame(titulo.groupby(by=['MUNICÍPIO','TITULAÇÃO'])['QTDE'].sum())
titulo.reset_index(inplace=True)
titulo=titulo.pivot(index='MUNICÍPIO', columns='TITULAÇÃO', values='QTDE')
titulo
```

```python
plotarHeatMap(titulo, titulo='Titulação de Docentes (Em exercício)', 
              xLabel='Titulação', yLabel='Município', cmap='Set3', lw=.5)
```
